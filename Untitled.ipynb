```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize = (20,5))
for i in range(10):
    plt.subplot(10, 1, i+1)
    plt.gray() 
    plt.imshow(digits.data[i].reshape(1,64)) 
    plt.axis('off')
# plt.gray() 
# plt.imshow(digits.data[0].reshape(1,64)) 

plt.show()
```
