```python
df_index2 = pd.DataFrame(columns = ["desc", "ref", "authors", "title", "journal"]+aa2)

for rec in aaindex.search(''):
    mat = np.zeros((len(aa), len(aa)))
    for i, a in enumerate(rec.rows.keys()):
        if a not in aa:
            rec.index.pop(i)
    for i, ind in enumerate(rec.index):
        mat[i, :i+1] = list(ind)
    mat[np.arange(len(aa)), np.arange(len(aa))]/=2
    mat = mat + mat.T
    df_index2.loc[rec.key] = [rec.desc, rec.ref, rec.authors, rec.title, rec.journal]+list(mat.flatten())
```
